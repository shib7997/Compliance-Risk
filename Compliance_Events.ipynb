```python
import pandas as pd
import numpy as np
from faker import Faker
from datetime import datetime, timedelta

fake = Faker()

rows = []

locations = ['US', 'CA', 'UK', 'RU', 'CN', 'DE']

for i in range(5000):

    login_attempts = np.random.randint(1, 15)
    failed_logins = np.random.randint(0, login_attempts)

    transaction_count = np.random.randint(1, 200)

    session_duration = np.random.randint(1, 120)

    location = np.random.choice(locations)

    timestamp = (
        datetime.now() -
        timedelta(minutes=np.random.randint(0, 100000))
    )

    rows.append({
        'user_id': fake.uuid4(),
        'login_attempts': login_attempts,
        'failed_logins': failed_logins,
        'transaction_count': transaction_count,
        'session_duration': session_duration,
        'access_location': location,
        'timestamp': timestamp
    })

compliance_data = pd.DataFrame(rows)

compliance_data.to_csv(
    'compliance_events.csv',
    index=False
)

print('Dataset created successfully')
```
